```python
# Load necessary libraries
import requests
from bs4 import BeautifulSoup
import os
```

```python
parents = soup.find("body").find("ol")

try:
    for link in parents:
        parent1 = list(link.descendants)
        parent1 = list(parent1)[0]['href']
        #print(parent1)


        next_page = requests.get(parent1)
        bs_page = BeautifulSoup(next_page.text, 'lxml')
        #print(bs_page.prettify())

        child_urls = bs_page.select("ol li a")
        for child_url in child_urls:
            url = child_url['href']
            #print(url)

            desire_page = requests.get(url)
            bs_desire_page = BeautifulSoup(desire_page.text, 'lxml')
            #print(bs_desire_page.prettify())    

            desire_urls = bs_desire_page.find("div", class_="entries-wrap")
            #print(desire_urls)

            #Grab the fOlder Name
            folder_name = bs_desire_page.select("div.breadcrumb a.taxonomy span")
            folder_name = folder_name[1].text
            #print(folder_name)

            # create a new folder
            if not os.path.exists("scraped_books/" + folder_name):
                os.makedirs("scraped_books/" + folder_name)


            #Grab the Book Name
            book_name = bs_desire_page.select("div.breadcrumb span.current-item")
            name = book_name[0].text.replace('/', '')
            print(name)

            with open("scraped_books/" + folder_name + "/" + name + ".txt", "w", encoding="utf-8") as f:
                for desire_url in desire_urls:
                    last_url = desire_url.select("article div.entry-wrap h2 a")
                    last_url = last_url[0]['href']
                    #print(last_url)

                    scraping_page = requests.get(last_url)
                    bs_scraping_page = BeautifulSoup(scraping_page.text, 'lxml')        

                    results = bs_scraping_page.select("div.entry-content p")
                    for res in results:
                        final_data = res.text
                        f.write(final_data + "\n")  # write the scraped data to the file
        
except Exception as e:
    print(e)
    
    
    
```

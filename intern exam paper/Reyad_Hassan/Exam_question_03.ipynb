Extract page start from: https://kathika.org/stories

```python

path = "C:/Users/Pappu/Desktop/books/chromedriver-win64/chromedriver.exe"
service = Service(path)

driver = webdriver.Chrome(service=service)

start_page= "https://kathika.org/stories?page="
story_content_page = "https://kathika.org/"


book_names = []
writer_names = []
book_links = []
book_contents = []

for i in range(1,7):
    stories_page = start_page + str(i)
    
    driver.get(stories_page)
    
    time.sleep(5)
        
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    stories = soup.find_all('div', class_='c-card')
    for story in stories:
        story_name = story.find('h3')
        book_names.append(story_name.text.strip())
        author_name = story.find('p').text.strip()
        writer_names.append(author_name)
        story_link = story.find('a',class_='')['href']
        book_links.append(story_link)
    driver.quit()
print(book_links)

# for i in range(len(book_links)):
#     print(len(book_links)-i+1)
#     url = "https://kathika.org" + book_links[i]
#     driver.get(url)
#     page_source = driver.page_source
#     soup = BeautifulSoup(page_source, 'html.parser')
#     content = soup.find('div', class_='story-content ql-editor')
#     desc = content.find_all('p')
#     string =""
#     for p in desc:
#         string = string + p.text.strip()
#         string = string.replace("\xa0","")
#         string = string.replace("\'","")
#         string = string.replace("'","")
#     book_contents.append(string)
```

Save each books as **txt** file. where file name is **book_name.txt**. inside that file is **book_content**.

Save each **book_name** and **writer_name** in **csv** file.

```python
#code:


# Saving books containing book_content as txt file
pattern_to_eliminate = r'[|?"\']'
for i in range(len(df1)):
    name = re.sub(pattern_to_eliminate, '',df1.loc[i].book_name)
    path = "books/" + name + ".txt"
    with open(path,'w') as file:
        file.write(str(df1.loc[i].book_content))

        
# Saving each book_name and writer_name is csv file
df2 = df1[['book_name','writer_name']]
df2.to_csv(r"book_name_and_writer.csv", index=False)
df2
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Calculate Budget for Each Division
average_budget = 2_000_000
divisions_budget = [average_budget * (1 + x) for x in [0.3, -0.5, 0.05, -0.07, -0.11, 0.5, -0.19, -0.02]]

# 2. Calculate Profit for Each Division for 3 Years
def calculate_profits(budget, years=3):
    profits = []
    for year in range(1, years + 1):
        profit_for_year = [budget * 0.5]
        budget *= (year + 1)  # Increase budget each year
        profits.append(profit_for_year)
    return profits

profits_per_division = []
for budget in divisions_budget:
    profits_per_division.append(calculate_profits(budget))

# 3. Create Lists for 8 Divisions and 3 Years of Profit
divisions = ['Division1', 'Division2', 'Division3', 'Division4', 'Division5', 'Division6', 'Division7', 'Division8']
year1_profits, year2_profits, year3_profits = [], [], []
for division_profits in profits_per_division:
    year1_profits.append(division_profits[0][0])
    year2_profits.append(division_profits[1][0])
    year3_profits.append(division_profits[2][0])

# 4. Create a Tuple from the Lists
data_tuple = (divisions, year1_profits, year2_profits, year3_profits)

# 5. Convert Tuple to Dictionary
data_dict = {
    'Divisions': data_tuple[0],
    'Year1': data_tuple[1],
    'Year2': data_tuple[2],
    'Year3': data_tuple[3]
}

# 6. Create a DataFrame
df = pd.DataFrame(data_dict).set_index('Divisions')

# 7. Visualize the DataFrame
df.plot(kind='barh', figsize=(12, 8))
plt.xlabel('Profit (in BDT)')
plt.ylabel('Divisions')
plt.title('Year wise Eight Divisions Profit (in BDT)')
plt.show()
```

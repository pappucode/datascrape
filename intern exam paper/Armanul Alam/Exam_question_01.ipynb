## Basic Data Structure Test

XYZ Company invested an average of 2 million BDT for each of its eight divisions in Bangladesh in the first year of operation for their new branches. The second and sixth divisions receive 30% and 50% more than the average budget, respectively. 1st, 3rd, 4th, 5th, 7th, and 8th divisions receive 50%, 5%, 7%, 11%, 19%, and 2% less funding than the average budget, accordingly. The budget for each division increases by double in the second year and by quadruple in the third year. Each year, each division's profit is half of the value of every year's budget. 

- #### Make 4 separate lists
	- One for 8 divisions
	- The remaining three for 3 years profits
- #### Make a tuple using the resulting 4 lists
- #### Convert the tuple to the dictionary 
- #### Make a data frame from the tuple and set index as Year1, Year2, Year3
- #### Visualize the data frame using horizontal bar plot
    - Label the both axis accordingly
    - Give the title as `Year wise Eight Divisions Profit(in BDT)`

### Start coding

```python
# Import Necessary libraries
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Code here for making 4 separate lists
# Hint: First, you have to find out budget, then profit
investment=float(2000000)
List_D=['Division1','Division2','Division3','Division4','Division5','Division6','Division7','Division8']
Y1_budget=[investment-(investment*(float(50/100))),investment+(investment*(float(30/100))),investment-(investment*(float(5/100))),
          investment-(investment*(float(7/100))),investment-(investment*(float(11/100))),investment+(investment*(float(50/100))),
          investment-(investment*(float(19/100))),investment-(investment*(float(2/100)))]
Y2_budget=[]
Y3_budget=[]
for i in Y1_budget:
    Y2_budget.append(2*i)
    Y3_budget.append(4*i)

List_p1=[]
List_p2=[]
List_p3=[]
for i in Y1_budget:
    List_p1.append(float(i/2))
    List_p2.append(float(float(2*i))/2)
    List_p3.append(float(float(4*i))/2)
    

# Print the results
print("List for Divisions:")
print(List_D)

print("\nList for 1st Year's Profit:")
print(List_p1)

print("\nList for 2nd Year's Profit:")
print(List_p2)

print("\nList for 3rd Year's Profits:")
print(List_p3)
```

Desired Output:

```python
# Code for tuple using the resulting list
result_tuple = tuple(zip(List_D, List_p1, List_p2, List_p3))
print("Tuple containing Divisions and 3 Years Profit:")
print(result_tuple)
```

Desired Output: 

```python
# Code for dictionary using the resulting tuple
result_dict = {i: [j, k, l]
               for i,j,k,l in zip(
                   List_D, List_p1, List_p2, List_p3
               )}

print("Dictionary containing Divisions and 3 Years Profit:\n")
print(result_dict)
```

Desired Output:

```python
p1 = {
    'Division1': 0.5,
    'Division2': 1.3,
    'Division3': 0.95,
    'Division4': 0.93,
    'Division5': 0.89,
    'Division6': 1.5,
    'Division7': 0.81,
    'Division8': 0.98
}


p2 = {division: budget * 2 for division, budget in p1.items()}
p3 = {division: budget * 4 for division, budget in p1.items()}

df = pd.DataFrame([p1, p2, p3], index=['Year1', 'Year2', 'Year3'])
df = df * 10**6
print(df)
```

Desired Output:

```python
# Visualize the data frame using horizontal bar plot
ax = df.plot(kind='barh', stacked=False)

plt.xlabel('Profit Amount (in BDT)')
plt.ylabel('Year')
plt.title('Year wise 8 Divisions Profit(in BDT)')
plt.legend(loc='lower right')
plt.show()
```
